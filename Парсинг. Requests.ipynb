```python
from bs4 import BeautifulSoup
import requests
import lxml.html as html
```

```python
url= "https://zooya.ru/sanktpeterburg/manchkin/" #страница, которую будет парсить
request = requests.get(url)
request #проверили статус запроса, пололжительный, можно использовать стандартные библиотеки
```

```python
request.text
```

```python
bs = BeautifulSoup(request.text, 'lxml') #или "html.parser")
```

```python
bs
```

```python
#название объявления
name = bs.find("div", class_ ="lh4").text
print (name)
```

```python
#Ссылка на объявление
link="https://zooya.ru"+ bs.find("div", class_ ="lh4").find("a").get("href")
link
```

```python
#цена
price = bs.find("div",class_="anim-info-price").text.replace('\n','')
price
```

```python
#дата публикации
data = bs.find("div", class_ = "anim-info-cust").find("label").text
data
```

```python
#кол-во объявлений
kot = bs.find_all("div", class_ = "anim-grid")
num = len(kot)
num
#всего нашлось 4 объявления
```

# найдем названия, цены, время публикации, ссылки на все объявления и сведем все в таблицу

```python
names=[]
prices=[]
links=[]
datas=[]
for cat in kot:
    name = cat.find("div", class_ ="lh4").text
    link="https://zooya.ru"+ cat.find("div", class_ ="lh4").find("a").get("href")
    price = cat.find("div",class_="anim-info-price").text.replace('\n','')
    data = cat.find("div", class_ = "anim-info-cust").find("label").text
    names.append(name)
    prices.append(price)
    links.append(link)
    datas.append(data)
data
pd.set_option('display.max_colwidth',150)
df=pd.DataFrame({'Наименование':names,'Цена':prices,'Время публикации':datas,'Ссылка на объявление':links})
df
```
